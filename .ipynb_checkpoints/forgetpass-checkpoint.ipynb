```python
from tkinter import *
from tkinter import messagebox
from PIL import ImageTk, Image
```

```python
import pymysql
```

```python
def login_page():
    window .destroy()
    %run LoginPage.ipynb
    
def clear():
    email_entry.delete(0,END)
    password_entry.delete(0,END)
    confirm_entry.delete(0,END)
    
def connect_database():
    if email_entry.get()=='' or password_entry.get()=='':
        messagebox.showerror('Error','All Fields Are Required')
    elif password_entry.get()!=confirm_entry.get():
        messagebox.showerror('Error','Password and Confirm Password does not Match')
    else:
        con=pymysql.connect(host='localhost',user='root',password='saja2001',database='user')
        mycursor=con.cursor()
        
        mycursor.execute('use user')
        query= 'select * from data where email=%s'
        mycursor.execute(query,(email_entry.get()))
            
        row=mycursor.fetchone()
        if row==None:
            messagebox.showerror('Error','Email is incorrect')
        else:
            mycursor.execute('use user')
            query= 'update data set password=%s where email=%s'
            mycursor.execute(query,(password_entry.get(),email_entry.get()))
            con.commit()
            con.close()
            messagebox.showinfo('Success','Password is reset, Please login with new password')
            clear()
            login_page()
            
    
def show():
        hide_button = Button(lgn_frame, image=hide_image, command=hide, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand2")
        hide_button.place(x=860, y=350)
        password_entry.config(show='')
        
def show_confirm():
        hide_confbutton = Button(lgn_frame, image=hide_confimage, command=hide_confirm, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand2")
        hide_confbutton.place(x=860, y=430)
        confirm_entry.config(show='')
        
def hide():
        show_button = Button(lgn_frame, image=show_image, command=show, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand2")
        show_button.place(x=860, y=350)
        password_entry.config(show='*')
        
        
def hide_confirm():
        show_confbutton = Button(lgn_frame, image=show_confimage, command=show_confirm, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand2")
        show_confbutton.place(x=860, y=430)
        confirm_entry.config(show='*')
        
# ========================================================================
# ============================GUI Code====================================
# ========================================================================        

window= Toplevel() #The Toplevel() function is used to create additional windows apart from the main Tkinter window.
window.geometry('1166x718') #sets the size
window.state('zoomed')
window.title('Forget Password Page')
          
# ========================================================================
# ============================background image============================
# ========================================================================
        
bg_frame = Image.open('background1.png')
photo = ImageTk.PhotoImage(bg_frame)
bg_panel = Label(window, image=photo)
bg_panel.image = photo
bg_panel.pack(fill='both', expand='yes')
        
# ====== Login Frame =========================
        
lgn_frame = Frame(window, bg='#040405', width=950, height=600)
lgn_frame.place(x=200, y=70)
        
# ========================================================================
# ========================================================================
        
txt = "Reset Password"
heading = Label(lgn_frame, text=txt, font=('yu gothic ui', 25, "bold"), bg="#040405",
                             fg='white',bd=5,relief=FLAT)
heading.place(x=80, y=30, width=300, height=30)
# ========================================================================
# ============ Left Side Image ================================================
# ========================================================================
        
side_image = Image.open('vector.png')
photo = ImageTk.PhotoImage(side_image)
side_image_label = Label(lgn_frame, image=photo, bg='#040405')
side_image_label.image = photo
side_image_label.place(x=5, y=100)
        
# ========================================================================
# ============ Password Image =============================================
# ========================================================================
        
sign_in_image = Image.open('reset2.png')
photo = ImageTk.PhotoImage(sign_in_image)
sign_in_image_label = Label(lgn_frame, image=photo, bg='#040405')
sign_in_image_label.image = photo
sign_in_image_label.place(x=640, y=60)

# ========================================================================
# ============================Email====================================
# ========================================================================
        
email_label = Label(lgn_frame, text="Email", bg="#040405", fg="#4f4e4d",
                                    font=("yu gothic ui", 13, "bold"))
email_label.place(x=550, y=250)

email_entry = Entry(lgn_frame, highlightthickness=0, relief=FLAT, bg="#040405", fg="#6b6a69",
                                    font=("yu gothic ui ", 12, "bold"), insertbackground = '#6b6a69')
email_entry.place(x=580, y=280, width=270)

email_line = Canvas(lgn_frame, width=300, height=2.0, bg="#bdb9b1", highlightthickness=0)
email_line.place(x=550, y=310)
        
# ===== Email icon =========
        
email_icon = Image.open('email3.png')
photo = ImageTk.PhotoImage(email_icon)
email_icon_label = Label(lgn_frame, image=photo, bg='#040405')
email_icon_label.image = photo
email_icon_label.place(x=550, y=280)

# ========================================================================
# ============================Reset button================================
# ========================================================================
        
reset_button = Image.open('btn1.png')
photo = ImageTk.PhotoImage(reset_button)
reset_button_label = Label(lgn_frame, image=photo, bg='#040405')
reset_button_label.image = photo
reset_button_label.place(x=550, y=480)
reset = Button(reset_button_label, text='Reset', font=("yu gothic ui", 13, "bold"), width=25, bd=0,
                            bg='#3047ff', cursor='hand2', activebackground='#3047ff', fg='white',command=connect_database)
reset.place(x=20, y=10)
        
# ========================================================================
# ============================password====================================
# ========================================================================
        
password_label = Label(lgn_frame, text="Password", bg="#040405", fg="#4f4e4d",
                                    font=("yu gothic ui", 13, "bold"))
password_label.place(x=550, y=320)

password_entry = Entry(lgn_frame, highlightthickness=0, relief=FLAT, bg="#040405", fg="#6b6a69",
                                    font=("yu gothic ui", 12, "bold"), show="*", insertbackground = '#6b6a69')
password_entry.place(x=580, y=350, width=244)

password_line = Canvas(lgn_frame, width=300, height=2.0, bg="#bdb9b1", highlightthickness=0)
password_line.place(x=550, y=380)
        
# ======== Password icon ================
password_icon = Image.open('password_icon.png')
photo = ImageTk.PhotoImage(password_icon)
password_icon_label = Label(lgn_frame, image=photo, bg='#040405')
password_icon_label.image = photo
password_icon_label.place(x=550, y=350)
        
# ========= show/hide password ==================================================================
        
show_image = ImageTk.PhotoImage \
            (file='show.png')

hide_image = ImageTk.PhotoImage \
            (file='hide.png')

hide_button = Button(lgn_frame, image=hide_image, command=hide, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand1")
hide_button.place(x=860, y=350)
        

# ========================================================================
# ============================confirm password====================================
# ========================================================================
        
confirm_label = Label(lgn_frame, text="Confirm Password", bg="#040405", fg="#4f4e4d",
                                    font=("yu gothic ui", 13, "bold"))
confirm_label.place(x=550, y=400)

confirm_entry = Entry(lgn_frame, highlightthickness=0, relief=FLAT, bg="#040405", fg="#6b6a69",
                                    font=("yu gothic ui", 12, "bold"), show="*", insertbackground = '#6b6a69')
confirm_entry.place(x=580, y=430, width=244)

confirm_line = Canvas(lgn_frame, width=300, height=2.0, bg="#bdb9b1", highlightthickness=0)
confirm_line.place(x=550, y=460)
        
# ======== confirm icon ================
conf_icon = Image.open('password_icon.png')
photo = ImageTk.PhotoImage(conf_icon)
conf_icon_label = Label(lgn_frame, image=photo, bg='#040405')
conf_icon_label.image = photo
conf_icon_label.place(x=550, y=430)
        
# ========= show/hide confirm password ==================================================================
        
show_confimage = ImageTk.PhotoImage \
            (file='show.png')

hide_confimage = ImageTk.PhotoImage \
            (file='hide.png')

hide_confbutton = Button(lgn_frame, image=hide_image, command=hide_confirm, relief=FLAT,
                                  activebackground="white"
                                  , borderwidth=0, background="white", cursor="hand1")
hide_confbutton.place(x=860, y=430)
        
        
        
   
        
       

      
        
        
        
window.mainloop()
```
